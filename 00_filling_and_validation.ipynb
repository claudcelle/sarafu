```python
users.area_name.value_counts()
```

```python
# 1. Kitui is a rural county. Since only 13 users are from Kitui, we decided to add them to other rural counties.
users_geo_areas = []
for i in users.index:
    area_name = users.area_name.loc[i]
    if area_name == 'Kitui':  
        users_geo_areas.append('Misc Rural Counties')
    if area_name != 'Kitui':
        users_geo_areas.append(area_name)
#
users['area_name'] = users_geo_areas
#
# 2. The name of Kakuma is substituted with the name of its county (Turkana).
users_geo_areas = []
for i in users.index:
    area_name = users.area_name.loc[i]
    if area_name == 'Kakuma':  
        users_geo_areas.append('Turkana')
    if area_name != 'Kakuma':
        users_geo_areas.append(area_name)
#
users['area_name'] = users_geo_areas
#
users.area_name.value_counts()
```
